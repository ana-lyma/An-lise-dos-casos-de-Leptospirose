#           Análise da Leptospirose pela faixa etária

```python
plt.title('Casos de Leptospirose por faixa etária em 2019')
plt.ylabel('Quantidade de Pessoas')
dfaixa2019.groupby('Faixa Etária')['Casos confirmados'].max().sort_values().plot(kind='bar',figsize=(10,4), color='#660033',rot=40)
```

```python
plt.title('Casos de Leptospirose por faixa etária em 2018')
plt.ylabel('Quantidade de Pessoas')
dfaixa2018.groupby('Faixa Etária')['Casos confirmados'].max().sort_values().plot(kind='bar',figsize=(10,4), color='#01A4A4',rot=40)
```

```python
plt.title('Casos de Leptospirose por faixa etária em 2017')
plt.ylabel('Quantidade de Pessoas')
dfaixa2017.groupby('Faixa Etária')['Casos confirmados'].max().sort_values().plot(kind='bar',figsize=(10,4),rot=40)
```

### Com a análise dos 3 graficos dos anos de 2017, 2018, e 2019, observa-se que a faixa predominante de casos é entre 20 à 59 anos
